```python
import pandas as pd
import os 
from pathlib import Path
from sqlalchemy import create_engine
import logging #provides a flexible and powerful way to track events that occur during the execution of a program
import time
logging.basicConfig(
    filename = "logs/ingestion_db.log",
    level = logging.DEBUG,
    format = "%(asctime)s - %(levelname)s - %(message)s",
    filemode = "a"
)
engine = create_engine('sqlite:///inventory.db')
folder =Path("C:\data")
print(folder.exists())
def ingest_db(df,table_name,engine):
    ''' this function will ingest the dataframe into  database table'''
    df.to_sql(table_name, con = engine,if_exists ='replace',index= False)
def load_raw_data():
    '''this function will load the csvs as dataframe and ingest into db'''
    start= time.time()
    for file in folder.glob("*.csv"):
        print(file.name)
        df = pd.read_csv(file)
        logging.info(f"ingesting {file} in db")
        ingest_db(df,file.stem,engine)#we use stem because file is a path object from patlib not a string so we cqnt use slicing
    end = time.time()
    total_time = (end - start)/60
    logging.info("ingestion complete")
    logging.info(f"\ntotal time taken:{total_time} minutes")    
if __name__=='__main__':
    load_raw_data()
                        
```
